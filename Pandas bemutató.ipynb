# Progress bar with tqdm

```python
import time
from tqdm import tqdm
```

```python
%%timeit
for i in tqdm(range(0,500)):
    time.sleep(0.001)
```

```python
start=time.time()
for i in (range(0,500)):
    time.sleep(0.001)
end=time.time()
print(end-start)
```
